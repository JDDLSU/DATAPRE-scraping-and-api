```python
import requests
import pandas as pd
import json
from bs4 import BeautifulSoup
```

```python
# get sources from NewsAPI
url = 'http://newsapi.org/v2/everything?'
api_key = 'bc22825921774073befe63f3a9247614'
sources = 'https://newsapi.org/v2/sources?apiKey=bc22825921774073befe63f3a9247614'
requests.get(sources).json()
```

**NOTE: The NewsAPI free plan only allows scraping from one month ago**

```python
# retrieve the articles from washingtonpost.com

parameters = {
	'domains' : 'inquirer.net',
	'apiKey' : api_key,
	'from' : '2022-04-12',
	'to' : '2022-05-12',
	'language' : 'en',
	'page' : '4'
}
```

```python
# store key value pairs in a variable

parameters_json = requests.get(url, params = parameters)
parameters_json = parameters_json.json()
parameters_json
```

```python
parameters_json.keys()
news_articles = parameters_json['articles']
news_articles
```

```python
article_results = []

for i in range (len(news_articles)):
	article = requests.get(news_articles[i]['url'])
	if(article.status_code == 200):
		soup = BeautifulSoup(article.content, 'html.parser')
		article_content = soup.find_all("div", {"class": "entry-content-wrap clearfix"})
		body_text = article_content[0].get_text()

		results_dict = {}
		results_dict['date'] = news_articles[i]['publishedAt']
		results_dict['title'] = news_articles[i]['title']
		results_dict['full_article'] = body_text
		results_dict['author'] = news_articles[i]['author']
	
		article_results.append(results_dict)
	
article_results
```

```python
import json

output = open("Dy_Velarde_api.json", "w", encoding='utf-8')

json.dump(article_results, output, indent=6, ensure_ascii=False)

output.close()
```
